```
from pyspark.sql import SparkSession
from pyspark.sql.types import *
import pyspark.sql.functions as F
```

```
spark = SparkSession \
    .builder \
    .appName("Wrangling Data") \
    .getOrCreate()
```

```
path = 'googleplaystore.csv'
```

```
play_schema = StructType([
                         StructField('App', StringType(), True),
                         StructField('Category', StringType(), True),
                         StructField('Rating', DoubleType(), True),
                         StructField('Reviews', IntegerType(), True),
                         StructField('Size', StringType(), True),
                         StructField('Installs', StringType(), True),
                         StructField('Type', StringType(), True),
                         StructField('Price', StringType(), True),
                         StructField('Content_Rating', StringType(), True),
                         StructField('Genres', StringType(), True),
                         StructField('Last_Updated', StringType(), True),
                         StructField('Current_Ver', StringType(), True),
                         StructField('Android_Ver', StringType(), True),
])
```

```
play_df = spark.read.csv(path, header=True, schema=play_schema)
```

```
play_df.printSchema()
```

```
play_df.show();
```

```
#drop type
play_df = play_df.drop('Type')
play_df.show()
```

```
#drop genre
play_df = play_df.drop('Genres')
play_df.show()
```

```
#add column
play_df.withColumn('app_name_upper', F.upper(play_df.App)).show()
```

```
#select category and ratings where reviews > 1000
play_df.select(['Category', 'Rating']).filter(play_df.Reviews > '1000').show()
```

```
#filter apps with category != BEAUTY
play_df.filter(play_df.Category != 'BEAUTY').show(250)
```

```
#select all apps with android version == 4.1 and up
play_df.filter(play_df.Android_Ver == '4.1 and up').show()
```

```
#select installs with rating < 3.0
play_df.select(['Installs']).filter(play_df.Rating < '3.0').show()
```
